```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 1锔 Cargar datos ya prefiltrados manualmente
df = pd.read_excel("base_limpia_grd_min20.xlsx")
df.columns = df.columns.str.strip()
df.fillna(df.mean(numeric_only=True), inplace=True)
df.fillna('Desconocido', inplace=True)

# 2锔 Features (verifica que estas columnas existan en tu archivo Excel)
features = [
    'Edad', 'Grupo Edad', 'Sexo', 'Codigo de ciudad', 'Tipo de ingreso',
    'D铆as estancia', 'ServicioAlta', 'Cuidados intensivos', 'D铆as de Unidad Cuidado Intensivo',
    'Dx de ingreso', 'Dx principal de egreso', 'Dx principal de egreso .1', 'Dx Ppal 3 Caracteres',
    'Dxr 1', 'Dxr 2', 'Dxr 3', 'Dxr 4', 'Dxr 5',
    'C贸digo causa externa', 'Causa externa', 'Situacion al alta',
    'Proc1', 'Proc2', 'Proc3', 'Proc4', 'Proc5', 'Proc6', 'Proc7', 'Proc8', 'Proc9', 'Proc10',
    'Tipo servicio', 'Causa Basica de muerte', 'Infecciones', 'Infecci贸n Quirurgica'
]

# Filtrar las features existentes en la base
features = [f for f in features if f in df.columns]

# 3锔 Matriz de entrada y etiquetas
X = pd.get_dummies(df[features], drop_first=True)
y = df['GRD -C贸digo'].astype(str)

# 4锔 Dividir antes de codificar etiquetas
X_train, X_test, y_train_raw, y_test_raw = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 5锔 Contar frecuencia total de GRD en todo el dataset
y_counts = y.value_counts()
valid_grds = y_counts[y_counts >= 20].index

# 6锔 Filtrar conjuntos de entrenamiento y prueba con solo GRD v谩lidos
mask_train = y_train_raw.isin(valid_grds)
mask_test = y_test_raw.isin(valid_grds)

X_train = X_train[mask_train]
y_train_raw = y_train_raw[mask_train]

X_test = X_test[mask_test]
y_test_raw = y_test_raw[mask_test]

# 7锔 Codificar etiquetas
le = LabelEncoder()
y_train = le.fit_transform(y_train_raw)
y_test = le.transform(y_test_raw)

# 8锔 Escalado
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

# 9锔 Entrenamiento con MLPClassifier
mlp = MLPClassifier(
    hidden_layer_sizes=(128, 64, 32),
    activation='relu',
    solver='adam',
    learning_rate_init=0.001,
    batch_size=32,
    max_iter=100,
    early_stopping=True,
    validation_fraction=0.1,
    random_state=42,
    verbose=True
)
mlp.fit(X_train_s, y_train)

#  Evaluaci贸n
y_pred = mlp.predict(X_test_s)
acc = accuracy_score(y_test, y_pred)

print(f"\n Accuracy: {acc:.4f}\n")
print(" Reporte de clasificaci贸n:")
print(classification_report(y_test, y_pred, target_names=le.inverse_transform(np.unique(y_test))))
print(" Matriz de confusi贸n:")
print(confusion_matrix(y_test, y_pred))
```
